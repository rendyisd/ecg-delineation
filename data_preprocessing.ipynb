## Extracting features and labels

```python
# LEADS = ['atr_avf', 'atr_avl', 'atr_avr', 'atr_i', 'atr_ii', 'atr_iii', 'atr_v1', 'atr_v2', 'atr_v3', 'atr_v4', 'atr_v5', 'atr_v6']
LEADS = ['atr_i', 'atr_ii', 'atr_iii']
longest_beat = 0

# Find the longest beat for zero padding
for ecg_index in range(1, 201):
    for lead in LEADS:
        s = ECGSignal.load_ecg_signal(ecg_index, lead)
        
        prev_p_start = None
        for seg, points in s.segment_start_end:
            start, end = points
            if prev_p_start is None and seg == 0:
                prev_p_start = start
            elif seg == 0:
                longest_beat = max(longest_beat, start - prev_p_start)

                if(longest_beat == start - prev_p_start):
                    temp = (ecg_index, lead)

                prev_p_start = start


print(longest_beat)
print(temp)
```

```python
CLASSES = [0, 1, 2, 3, 4, 5, 6]

X = []
y = []

for record_num in range(1, 201):
    for lead in LEADS:
        try:
            s = ECGSignal.load_ecg_signal(record_num, lead)

            for beat in s.cut_per_beat():
                signal, segment_map = beat
                signal = np.array(signal)
                segment_map = np.array(segment_map)
                segment_map = ECGSignal.one_hot_segment_map(segment_map, CLASSES)

                # Zero padding
                pad_length = longest_beat - len(signal)

                signal = np.pad(signal, (0, pad_length), mode='constant', constant_values=0)
                segment_map = np.pad(segment_map, ((0, pad_length), (0, 0)), mode='constant', constant_values=0)

                X.append(signal)
                y.append(segment_map)

        except:
            # Print failed instance
            print(f"Record: {record_num} | Lead: {lead}", end=' , ')
```

```python
X = np.array(X)
y = np.array(y)

# Reshape X
X = X.reshape(X.shape[0], X.shape[1], 1)

X.shape, y.shape
```
